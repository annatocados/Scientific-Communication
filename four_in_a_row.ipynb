```python
#Variable del joc

Espai_buit = " "
 

#Variable global

M =[]

 

#Aquesta funció és per crear la matriu de la dimensió determinada, que tindrà com a input la variable global definida anteriorment, és a dir, una matriu buida, a la qual establim que hi hagin espais buits com a elements, per despres omplir-los amb les diferents fitxes

def crear_matriu():

   

    for n in range (6):

        M.append ([])

        for m in range (7):

            M[n].append(Espai_buit)

 

#Mostrem la matriu amb la següent funcio, abans l'hem creat amb una certa dimensió però aquesta funció servirà per establir el disseny i la impressio de la nostra matriu. En el nostre cas perquè les caselles estiguin buides per dins i les parets estiguin delimitades i així obtingui un disseny semblant al d'una matriu.
#Com a input té la matriu creada amb la funció anterior
def mostrar_tauler(M):

    for n in range(6):

        print ("|", end="")

        for m in range(7):

            print (M[n][m], end="")

            print ("|", end="")

        print ("")

 

#Aquesta funció té com a objectiu que el jugador esculli on col·locar la fitxa entre les 7 columnes que hi ha, per tant, la fitxa es col·locarà a la menor fila possible que estigui lliure.

def jugador_funcio ():

  

    jugador = int(input("A quina columna vols posar la fitxa? "))

    n = 5

  

    while n >= 0:

      

        if M[n][jugador] == " ":

           

            M[n][jugador] = "O"

         

            break 

       

        n = n - 1

      

 

#Jugada del ordinador, en aquest cas s'escull una columnar totalment a l'atzar, i igual que l'anterior funció, la fitxa es col·locarà a la menor fila lliure possible.

def random_jugada ():

    import random

    a = random.randint(0,6)

    n = 5

  

    while n > 0:

      

        if M[n][a] == " ":

          

            M[n][a]="X"

          

            break

      

        n = n - 1

 

#QUATRE EN RATLLA. Les següents funcions són per detectar el 4 en ratlla i per indicar si s'ha guanyat o no el joc. Això ho fem comprovant que el caràcter d'un espai de la matriu és igual al següent (durant 4 vegades). 

 

def quatre_horitzontal ():

    for n in range (0, 6):

        m = 0

        while m <=3:

            if M[n][m]==M[n][m+1]==M[n][m+2]==M[n][m+3]=="X": 

                return True

            if M[n][m]==M[n][m+1]==M[n][m+2]==M[n][m+3]=="O":

                return True

            m = m + 1

  

def quatre_vertical ():

    for m in range (0,7):

        n = 0

        while n <=2:

            if M[n][m]==M[n+1][m]==M[n+2][m]==M[n+3][m]=="X":

                return True

            if M[n][m]==M[n+1][m]==M[n+2][m]==M[n+3][m]=="O":

                return True

            n = n + 1

 

def diagonal_positiva():

    for n in range (0, 6):

        m = 0 

        while m <= 3 and n <= 2:

            if M[n][m]==M[n+1][m+1]==M[n+2][m+2]==M[n+3][m+3]=="X":

                return True

            if M[n][m]==M[n+1][m+1]==M[n+2][m+2]==M[n+3][m+3]=="O":

                return True

          

            m = m + 1

          

def diagonal_negativa():

    for n in range (0, 6):

        m = 6

        while m >=3 and n <=2:

            if M[n][m]==M[n+1][m-1]==M[n+2][m-2]==M[n+3][m-3]=="X":

                return True

            if M[n][m]==M[n+1][m-1]==M[n+2][m-2]==M[n+3][m-3]=="O":

                return True

          

            m = m - 1


#Dins de les funcions de detecció de quatre en ratlla, aquesta es la que les ajunta totes per poder indicar quan s'acaba el joc.

def comprovar ():

    if quatre_horitzontal()==True or quatre_vertical()==True or diagonal_positiva()==True or diagonal_negativa()==True:

        return True


    else:

        return False
    


#Cridar les diferents funcions plantejades.

def main ():

   

    V = 0 #Amb això predeterminem que comenci el joc sent la funció comrpovar() fals.

   

    crear_matriu()
    



    while (V == 0):

       

        jugador_funcio()

           

        if (comprovar () == True):

 

            print("Has guanyat!")

           

            V = 1 #Quan has guanyat, la funció  es canvia la variable a 1 per no tornar a començar el bucle.

       

        else:

           

            random_jugada()

           

            if (comprovar () == True):

               

                print("Has perdut!")

               

                V = 1

       

        mostrar_tauler(M)

       

if __name__ == '__main__':

     main()
     
     
#El joc de proves que hem fet per comprobar que el programa funciona amb la seva totalitat ha sigut el següent:
    #Per una part, hem provat que el jugador pugui guanyar utilitzant les diferents funcions que hem plantejat, és a dir, que pugui guanyar amb un connecta 4 vertical, horitzontal, diagonal positiva i diagonal negativa. A la vegada hem provat que l'ordinador també pugui guanyar utilitzant les 4 estratègies plantejades anteriorment(vertical,hortizontal,diagonal positiva,diagonal negativa)
```
