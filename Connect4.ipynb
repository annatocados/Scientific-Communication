```python
#Game variable 

Espai_buit = " "
 

#global variable 

M =[]

 

#This function is used to create a matrix of the specified dimension, which will take as input the previously defined global variable, i.e., an empty matrix, in which we establish that there are empty spaces as elements, and then fill them with different tokens.
def crear_matriu():

    for n in range (6):

        M.append ([])

        for m in range (7):

            M[n].append(Espai_buit)

 

#We display the matrix with the following function, that we previously created with a certain dimension, but this function will serve to establish the design and printing of our matrix. In our case, we want the cells to be empty inside and the walls to be delimited, so that it resembles the design of a matrix.
#As input, it takes the matrix created with the previous function.
def mostrar_tauler(M):

    for n in range(6):

        print ("|", end="")

        for m in range(7):

            print (M[n][m], end="")

            print ("|", end="")

        print ("")

 

#The objective of this function is for the player to choose where to place their token among the 7 columns available. Therefore, the token will be placed in the lowest possible row that is available.
def jugador_funcio ():

    jugador = int(input("What column do you want to place the token in?")) #You have to take into account that the first column is number 0!

    n = 5

    while n >= 0:


        if M[n][jugador] == " ":

           

            M[n][jugador] = "O"

         

            break 
    

        n = n - 1
 

#Computer's turn, in this case, a column is chosen completely at random, and just like the previous function, the token will be placed in the lowest possible free row.
def random_jugada ():

    import random

    a = random.randint(0,6)

    n = 5

    while n > 0:

        if M[n][a] == " ":

          

            M[n][a]="X"

          

            break

      

        n = n - 1

 

#CONNECT FOUR. The following functions are for detecting a connect four and indicating whether the game has been won or not. This is done by checking if the character in a space of the matrix is equal to the next one (repeated four times).

def quatre_horitzontal ():

    for n in range (0, 6):

        m = 0

        while m <=3:

            if M[n][m]==M[n][m+1]==M[n][m+2]==M[n][m+3]=="X": 

                return True

            if M[n][m]==M[n][m+1]==M[n][m+2]==M[n][m+3]=="O":

                return True

            m = m + 1

  

def quatre_vertical ():

    for m in range (0,7):

        n = 0

        while n <=2:

            if M[n][m]==M[n+1][m]==M[n+2][m]==M[n+3][m]=="X":

                return True

            if M[n][m]==M[n+1][m]==M[n+2][m]==M[n+3][m]=="O":

                return True

            n = n + 1

 

def diagonal_positiva():

    for n in range (0, 6):

        m = 0 

        while m <= 3 and n <= 2:

            if M[n][m]==M[n+1][m+1]==M[n+2][m+2]==M[n+3][m+3]=="X":

                return True

            if M[n][m]==M[n+1][m+1]==M[n+2][m+2]==M[n+3][m+3]=="O":

                return True

          

            m = m + 1

          

def diagonal_negativa():

    for n in range (0, 6):

        m = 6

        while m >=3 and n <=2:

            if M[n][m]==M[n+1][m-1]==M[n+2][m-2]==M[n+3][m-3]=="X":

                return True

            if M[n][m]==M[n+1][m-1]==M[n+2][m-2]==M[n+3][m-3]=="O":

                return True

          

            m = m - 1


#Within the connect four detection functions, this is the one that combines them all to indicate when the game is over.
def comprovar ():

    if quatre_horitzontal()==True or quatre_vertical()==True or diagonal_positiva()==True or diagonal_negativa()==True:

        return True


    else:

        return False
    


#Call the different functions we propose. 

def main ():

   

    V = 0 #Amb això predeterminem que comenci el joc sent la funció comrpovar() fals.

   

    crear_matriu()
    



    while (V == 0):

       

        jugador_funcio()

           

        if (comprovar () == True):

 

            print("You won!")

           

            V = 1 #When you win, the function changes the variable to 1 so the bucle doesn't start all over again. 

       

        else:

           

            random_jugada()

           

            if (comprovar () == True):

               

                print("You lost!")

               

                V = 1

       

        mostrar_tauler(M)

       

if __name__ == '__main__':

     main()
     
     
#The test game we have conducted to verify that the program works in its entirety is as follows:
    #On one hand, we have tested that the player can win using the different functions we have designed, meaning they can win with a vertical, horizontal, positive diagonal, and negative diagonal connect four. At the same time, we have also tested that the computer can win using the four strategies mentioned before (vertical, horizontal, positive diagonal, negative diagonal).
```
